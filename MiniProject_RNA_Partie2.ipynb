<h3 style=" color : #FF3131">Partie 2 : Elbow methode </h3>

<h4 style="color:#702963">Introduction : </h4>

<p>Dans l'analyse de cluster, la méthode Elbow est une heuristique utilisée pour déterminer le nombre de clusters dans un ensemble de données. La méthode consiste à tracer la variation expliquée en fonction du nombre de clusters, et à choisir le coude de la courbe comme nombre de clusters à utiliser.</p>

<h4 style="color:#702963">Lecture de fichier : </h4>

<ul>
    <li>
    L'importation des bibliothèques : 
    </li>
</ul>


````python
%%capture
!pip install sklearn matplotlib seaborn
````

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
````

<ul>
    <li>
    Nettoyer les données : 
    </li>
</ul>


````python
# saisir le chemin du ficher csv
filepath = 'EducationSéries2005-2010.csv'

# creer le dataframe depuis le fichier csv dèja créer
df = pd.read_csv(filepath)
# supprimer les lignes doublants
df.drop_duplicates(inplace=True)

#remplacer les cases non définie NaN
df = df.fillna(0)
````

<ul>
    <li>
    Affichage de quelque donnée : 
    </li>
</ul>

````python
print(df.head())

print(df.iloc[:,3:])
````

<h4 style="color:#702963">Implementation de methode Elbow : </h4>

````python
# on va focaliser sur seulement les valeurs de l'année 2005 à 2010
X = df.iloc[:,3:].values
# array pour ajouter inertia
wcss = []
# la boucle de 1 à 11 clustering
for i in range(1, 11):
    # appel de Kmeans
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
````

<ul>
    <li>
    L'affichage de résultat : 
    </li>
</ul>

````python
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
````

Comme vous vouyez, le coude ( Elbow ) de graphe est appoché de numero deux, donc on va prendre deux clusters.

<h3 style=" color : #FF3131">Partie 3 : Application de modele </h3>

<h4 style="color:#702963">Introduction : </h4>

<p>SimpSOM est une implémentation légère de Kohonen Self-Organizing Maps (SOM) pour Python 2.7 et 3, utile pour l'apprentissage non supervisé, le clustering et la réduction de dimensionnalité. Il vous permet de créer et d'entraîner le SOM sur votre ensemble de données, d'enregistrer/charger les poids de réseau entraînés et d'afficher ou d'imprimer des graphiques du réseau avec des fonctionnalités sélectionnées. 
</p>

<h4 style="color:#702963">Implementation de methode : </h4>

<ul>
    <li>
    L'instalation des biblithèques : 
    </li>
</ul>

````python
%%capture
!pip install numpy pandas sklearn SimpSOM matplotlib plotly
````

````python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
````

<ul>
    <li>
    Lecture de fichier : 
    </li>
</ul>

````python
#Lire CSV
data = pd.read_csv("EducationSéries2005-2010.csv")

#Enregistrer les pays pour la visualisation
countries = data.iloc[:,1].values

#conserver seulement les données numériques
data = data.drop(data.columns[0:3],axis=1)
data = data.fillna(0)
````

<ul>
    <li>
    Normalisation des données : 
    </li>
</ul>

````python
#normaliser les valeurs
train = StandardScaler().fit_transform(data.values)
#affichage
print(pd.DataFrame(train).head())
print(np.unique(countries))
````

<ul>
    <li>
    L'utilisation de SimpSom : 
    </li>
</ul>

````python
#Import the library
import SimpSOM as sps

#Build a network 12x12 with a weights format taken from the raw_data and activate Periodic Boundary Conditions. 
net = sps.somNet(12, 12, train)

#Train the network for 10000 epochs and with initial learning rate of 0.01. 
net.train(0.01, 10000)

#Save the weights to file
net.save('som_weights_Education')
````

<ul>
    <li>
    Les clusters : 
    </li>
</ul>

````python
clusters = net.cluster(train,show=True,type="KMeans",numcl=2)
````

<ul>
    <li>
    On voit le shape de notre données d'entrainement et le nombre de clusters finaux : 
    </li>
</ul>

````python
print(train.shape)

#le nombre de clusters
n = np.size(clusters)

print(n)
````

<ul>
    <li>
    On affiche les indices des pays existe dans chaque clusters : 
    </li>
</ul>

````python
CountryClusters = list()
for i in range(n):
    CountryClusters.append(countries[clusters[i]])
````

<ul>
    <li>
    le nombre des données dans chaque clusters : 
    </li>
</ul>

````python
for idx,c in enumerate(CountryClusters):
    print("COUNTRIES OF {} CLUSTER : {}".format(idx,len(c)))
````

<ul>
    <li>
    Projeté nos données sur la carte auto-organisatrice : 
    </li>
</ul>

````python
import matplotlib.pyplot as plt
project = np.array(net.project(train),dtype='float')
````

<b>La méthode <code>project()</code> nous permet d'avoir le positionnement de chaque BMU ( best matching unit) ou bien le neuronne gangant pour chaque ligne d'entrée, cela nous sera utile lors de l'affichage des noms des pays</b>

````python
x = project[:,0]
y = project[:,1]
````

````python
#on creer une liste qui continet le classification de chaque pays
labels = np.zeros(countries.size)
for k,c in enumerate(clusters):
    for i in c:
        labels[i] = k
````

````python
print(labels)
````

#### Pour éviter un regroupement illisible :
- On regroupe tous les neuronnes qui classifie la même pays en une seule cellule (unit) en calculant la moyennes de leurs coordonnées x et y
- On classifie ces cellules en choisisant la classification la plus pertinente dans tous les neuronnes

#### Définition des variables utilisé :
- x : coordonnés selon l'axe des x
- y : coordonnés selon l'axe des y
- labels : la classification de chaque entrés des données d'entrainement <code>train</code>
- clist : Noms des pays distinct ( sans répétition ), pour éviter le calcule innécessaire
- countries : liste des noms des pays pour chaque ligne d'entré

````python
x = np.array(x)
y = np.array(y)
labels = np.array(labels)

compressed_points = []
clist = []
for c in countries:
    if c not in clist:
        clist.append(c)
        #chercher les index du pays c
        idx = np.where(countries == c)
        idx = idx[0]
        
        #calculer les moyennes
        moyX = np.sum(x[idx])/len(idx)
        moyY = np.sum(y[idx])/len(idx)
        
        #Chercher le classement le plus fréquent
        unique, counts = np.unique(labels[idx], return_counts=True)
        maxLabel = unique[np.argmax(counts)]

        #ajouter les données dans une seule ligne
        compressed_points.append([moyX,moyY,int(maxLabel),c])
````

````python
compressed_points = np.array(compressed_points)

print(compressed_points[:5,:])
````

<ul>
    <li>
    Préparation des données pour l'affichage : 
    </li>
</ul>

````python
x = np.array(compressed_points[:,0],dtype='float')
y = np.array(compressed_points[:,1],dtype='float')
labels = np.array(compressed_points[:,2],dtype=int)
countryNames = compressed_points[:,3]

print(" x : {} ,  y : {} , labels : {} , countryNames : {} ".format(x[0],y[0],labels[0],countryNames[0]))
````

<ul>
    <li>
    L'affichage des resultat avec nom de pays : 
    </li>
</ul>

````python
#x et y pour le graphe
scatter_x = np.array(x)
scatter_y = np.array(y)

#classification des pays
group = np.array(labels,dtype=int)

#couleurs pour chaque cluster
cdict = {1: 'red', 2: 'blue', 3: 'green'}

#Configuration du graphe
plt.rcParams["figure.figsize"] = (10,10)
fig, ax = plt.subplots()
for g in np.unique(group):
    i = np.where(group == g)
    ax.scatter(scatter_x[i], scatter_y[i], label=g)
ax.legend()

#ajouter le nom du pays pour chaque point
for i, txt in enumerate(countryNames):
    ax.annotate(txt, (x[i], y[i]))

#affichage du graphe    
plt.show()
````

<ul>
    <li>
    Une autre forme d'affichage avec zoom sur chaque pays :
